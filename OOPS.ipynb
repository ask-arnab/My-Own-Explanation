```python
# If a function is defined inside of a class, then it is called a method
```

```python
class Atm:
    def __init__(self): # -> Constructor of the class, and it got a superpower. (Magic Method)
        #print(id(self))
        self.pin = ''
        self.balance = 0
        self.menu()
    def menu(self):
        user_input = int(input("""
        Press 1 for create pin
        Press 2 for change pin
        Press 3 for check balance
        press 4 for withdraw
        Press any key for exit
        """))
        if user_input ==1:
            self.create_pin()
        elif user_input == 2:
            self.change_pin()
        elif user_input ==3:
            self.check_balance()
        elif user_input ==4:
            self.withdraw()
        else:
            exit()
    def create_pin(self):
        pin = input("Enter Your Pin")
        self.pin = pin
        balance = float(input("Enter your balance"))
        self.balance = balance
        print("Pin created Successfully")
        self.menu()
    def change_pin(self):
        current_pin = input("Enter your current pin")
        if current_pin == self.pin:
            new_pin = input("Enter your new pin")
            self.pin = new_pin
            print("Pin change successfully")
        else:
            print("Esob churi chamari ekahne cholbe na")
        self.menu()
    def check_balance(self):
        current_pin = input("Enter your current pin")
        if current_pin == self.pin:
            print(f"your current balance is {self.balance}")
        else:
            print("Pin is invalid")
        self.menu()
    def withdraw(self):
        current_pin = input("Enter your current pin")
        if current_pin == self.pin:
            amount = float(input("Enter your amount"))
            if amount <= self.balance:
                self.balance = self.balance - amount
                print("withdraw successful")
                print("current balance is",self.balance)
            else:
                print("garib")
        else:
            print("TMC")
        self.menu()
```

```python
atm = Atm()
print(id(atm))
```
