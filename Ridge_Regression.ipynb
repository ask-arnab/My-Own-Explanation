```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("Huber_Loss.csv")
```

```python
X = df.iloc[:,0].values
Y = df.iloc[:,1].values
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=2)
```

```python
plt.scatter(X_train,Y_train)
```

```python
from sklearn.metrics import r2_score
from sklearn.linear_model import SGDRegressor
```

```python
sgd = SGDRegressor(loss = 'squared_error',penalty=None,max_iter=100,eta0=0.0001,learning_rate='constant',early_stopping=True)
```

```python
sgd.fit(X_train.reshape(-1,1),Y_train)
```

```python
r2 = r2_score(Y_test,sgd.predict(X_test.reshape(-1,1)))
print(r2)
```

```python
r2 = r2_score(Y_train,sgd.predict(X_train.reshape(-1,1)))
print(r2)
```

```python
plt.scatter(X_test,Y_test)
plt.plot(X_test,sgd.predict(X_test.reshape(-1,1)),color='red')
```

```python
#Using regularization
```

```python
sgd_ridge = SGDRegressor(loss = 'squared_error',
                         penalty='l2',
                         alpha=0.0003,
                         l1_ratio=0,
                         max_iter=100,
                         eta0=0.0001,
                         learning_rate='constant',
                         early_stopping=True)
```

```python
sgd_ridge.fit(X_train.reshape(-1,1),Y_train)
```

```python
r2 = r2_score(Y_test,sgd_ridge.predict(X_test.reshape(-1,1)))
print(r2)
```

```python
r2 = r2_score(Y_train,sgd_ridge.predict(X_train.reshape(-1,1)))
print(r2)
```

```python
plt.scatter(X_test,Y_test)
plt.plot(X_test,sgd_ridge.predict(X_test.reshape(-1,1)),color='red')
```

```python
sgd_huber = SGDRegressor(loss = 'huber',
                         penalty=None,
                         max_iter=100,
                         eta0=0.0001,
                         learning_rate='constant',
                         early_stopping=True)
```

```python
sgd_huber.fit(X_train.reshape(-1,1),Y_train)
```

```python
r2 = r2_score(Y_test,sgd_huber.predict(X_test.reshape(-1,1)))
print(r2)
```
