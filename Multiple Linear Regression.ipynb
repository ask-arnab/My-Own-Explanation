```python
from sklearn.datasets import load_diabetes
```

```python
import pandas as pd
import numpy as np
```

```python
X,Y = load_diabetes(return_X_y=True,as_frame=True)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=5)
```

```python
#Multiple Linear Regreesion
```

```python
# y = beta1*x1+beta2*x2+beta3*x3+......+betan*xn+beta0 , beta = [0,0,0,0,0,0,...,0]
```

```python
class MyLinearRegression:
    #Constructor
    def __init__(self):
        self.intercept = None
        self.coef = None
    #Train
    def train(self,x_train,y_train):
        self.intercept = 0
        self.coef = np.zeros(x_train.shape[1])
        x_train = np.insert(x_train,0,1,axis=1)
        inv_part = np.linalg.inv(np.dot(x_train.T,x_train))
        beta = np.dot(inv_part,np.dot(x_train.T,y_train))
        self.intercept = beta[0]
        self.coef = beta[1:]
        return(self.intercept,self.coef)
        
```

```python
mlr = MyLinearRegression()
```

```python
mlr.train(X_train,Y_train)
```

```python
#Mentos Zindegi
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train,Y_train)
```

```python
lr.intercept_
```

```python
lr.coef_
```

```python
from sklearn.metrics import mean_squared_error,r2_score
```

```python
mse = mean_squared_error(Y_test,lr.predict(X_test))
print(np.sqrt(mse))
```

```python
r2 = r2_score(Y_test,lr.predict(X_test))
print(r2)
```
